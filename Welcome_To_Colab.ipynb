With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

Identify all patterns within each column. Determine which frequencies dominate in high versus low attention states. ## Load the data

### Subtask:
Load the dataset into a pandas DataFrame.


**Reasoning**:

```
import pandas as pd
df = pd.read_csv('/content/acquiredDataset.csv')
display(df.head())
```

## Analyze patterns in each column

### Subtask:
Analyze the distribution and characteristics of each column in your dataset.

**Reasoning**:

```
import matplotlib.pyplot as plt

display(df.describe())

df.hist(figsize=(15, 10), bins=20)
plt.tight_layout()
plt.show()
```

## Investigate frequency dominance

### Subtask:
Investigate the relationship between attention levels and the different frequency bands. We will look at which frequencies are more prominent when attention is high versus when it is low.


**Reasoning**:
I will now divide the dataset into two groups based on the median of the 'attention' column, calculate the mean of each frequency band for both groups, and then create bar plots to visualize and compare these means, as instructed.



```
attention_median = df['attention'].median()
high_attention = df[df['attention'] > attention_median]
low_attention = df[df['attention'] <= attention_median]

frequency_bands = ["delta", "theta", "lowAlpha", "highAlpha", "lowBeta", "highBeta", "lowGamma", "highGamma"]

high_attention_means = high_attention[frequency_bands].mean()
low_attention_means = low_attention[frequency_bands].mean()

comparison_df = pd.DataFrame({'High Attention': high_attention_means, 'Low Attention': low_attention_means})



comparison_df.plot(kind='bar', figsize=(12, 7))
plt.title('Mean Frequency Band Power in High vs. Low Attention States')
plt.ylabel('Mean Power')
plt.xlabel('Frequency Bands')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Analyze frequency shifts with classification

### Subtask:
Analyze how the power of each frequency band changes depending on the 'classification' value (0 or 1).


**Reasoning**:
To analyze how the power of each frequency band changes with the 'classification' value, I will first filter the dataframe into two separate dataframes based on 'classification' being 0 or 1. Then, for each of these dataframes, I will calculate the mean of the frequency band columns. I will then combine these means into a new dataframe and visualize the comparison using a bar plot, ensuring the plot is well-labeled and readable.



```
classification_0 = df[df['classification'] == 0]
classification_1 = df[df['classification'] == 1]

frequency_bands = ["delta", "theta", "lowAlpha", "highAlpha", "lowBeta", "highBeta", "lowGamma", "highGamma"]

classification_0_means = classification_0[frequency_bands].mean()
classification_1_means = classification_1[frequency_bands].mean()

comparison_classification_df = pd.DataFrame({'Classification 0': classification_0_means, 'Classification 1': classification_1_means})

comparison_classification_df.plot(kind='bar', figsize=(12, 7))
plt.title('Mean Frequency Band Power for Classification 0 vs. 1')
plt.ylabel('Mean Power')
plt.xlabel('Frequency Bands')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

*   There is a strong relationship between the "classification" variable and brainwave power. The mean power for all measured frequency bands is significantly higher for `classification == 0` than for `classification == 1`.
